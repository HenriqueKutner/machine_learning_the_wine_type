```python
import pandas as pd
arquivo = pd.read_csv(r'C:\Users\HKC_H\OneDrive - pluus20\HKC\data-analytics\Learn\machineLearningProjectWine\wine_dataset.csv')
arquivo
```

```python
arquivo.head()
```

```python
arquivo['style'] = arquivo['style'].replace('red', 0)
```

```python
arquivo['style'] = arquivo['style'].replace('white', 1)
```

```python
arquivo
```

```python
# Separando as variáveis entre preditoras e variável alvo.

y = arquivo['style'] # Selecionei a variável alvo
x = arquivo.drop('style', axis = 1) # Declarei todas as variáveis preditoras e tirei a coluna style com o comando drop
```

```python
# Função de treino

from sklearn.model_selection import train_test_split

# Criando os conjuntos de dados de treino e teste

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3) 

# o 0.3 means that we want just 30% of the date for tests e the anpther 70% will be for training
# Da função a direita irá retornar 4 saídas que estão a esquerda
```

```python
from sklearn.ensemble import ExtraTreesClassifier
#Criação do modelo

modelo = ExtraTreesClassifier()
modelo.fit(x_treino, y_treino)

#Imprimindo resultados

resultado = modelo.score(x_teste, y_teste)

# Para x é passado os valores preditores e com esses dados ele vai tentar prever o resultado
# E comparar com o y_teste se ele acertou ou não com isso calculando a accuracy
print("Acurácia:", resultado)
```
